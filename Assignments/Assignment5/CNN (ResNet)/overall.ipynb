```python
tumorDataset = TumorDataset()
plt.figure()
plt.imshow(tumorDataset[10][0])
```

```python
from torch.utils.data import random_split

train_ratio = 0.7
validation_ratio = 0.1

train_size = int(train_ratio * len(tumorDataset))
validation_size = int(validation_ratio * len(tumorDataset))
test_size = len(tumorDataset) - (train_size + validation_size)

train_data, validation_data, test_data = random_split(tumorDataset, [train_size, validation_size, test_size])

train_loader = DataLoader(train_data, batch_size=256, shuffle=True, num_workers=1)
validation_loader = DataLoader(validation_data, batch_size=1, shuffle=True, num_workers=1)
test_loader = DataLoader(test_data, batch_size=1, shuffle=True, num_workers=1)
```

```python
import torch
# from tqdm.notebook import tqdm
from torch import optim
import torch.nn as nn

def train(net, epochs):
    global train_loader, validation_loader
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(net.parameters(), lr=0.001)

    loss_record = {
        "train": [], "validation": []
    }
    for epoch in range(epochs):
        net.train()

        running_loss = 0
        for i, data in enumerate(train_loader):
            images, labels = data
            images = torch.tensor(images)[:, None, :, :].float().cuda()
            labels = torch.tensor(labels).long().cuda()

            # print(labels.sum())
            optimizer.zero_grad()
            outputs = net(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()
            if i % 10 == 9:    # print every 2000 mini-batches
                print('[%d, %5d] loss: %.3f' %
                      (epoch + 1, i + 1, running_loss / 10))
                running_loss = 0.0
            
        
        net.eval()
        validation_loss = 0
        with torch.no_grad():
            for i, data in enumerate(validation_loader):
                images, labels = data
                images = torch.tensor(images)[:, None, :, :].float().cuda()
                labels = torch.tensor(labels).long().cuda()
                output = net(images)
                loss = criterion(output, labels)
                validation_loss += loss.item() * len(images)
        
        train_loss = 0
        with torch.no_grad():
            for i, data in enumerate(train_loader):
                images, labels = data
                images = torch.tensor(images)[:, None, :, :].float().cuda()
                labels = torch.tensor(labels).long().cuda()
                output = net(images)
                loss = criterion(output, labels)
                train_loss += loss.item() * len(images)

        print( "[epoch : %4d] training loss: %.4f, validation loss: %.4f" % (epoch, train_loss / (len(train_loader) * 256), validation_loss / len(validation_loader)))
        loss_record['train'].append(train_loss / (len(train_loader) * 256))
        loss_record['validation'].append(validation_loss / len(validation_loader))
    
    return loss_record
```

```python
from resnet import ResNet

net = ResNet()
device = torch.device("cuda:0")
net.to(device)
record = train(net, 3)
```

```python
plt.plot(record["train"], label = "train loss")
plt.plot(record["validation"], label = "validation loss")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import accuracy_score


def test(net):
    net.eval()

    sum_accuracy = 0
    with torch.no_grad():
      for i, data in enumerate(test_loader):
            images, labels = data
            images = torch.tensor(images)[:, None, :, :].float().cuda()
            labels = torch.tensor(labels).long().cuda()
            outputs = net(images)
            preds = torch.argmax(outputs, 1).cpu().numpy()
            sum_accuracy += accuracy_score(labels.cpu().numpy(), preds)

    print("accuracy: ", sum_accuracy / len(test_loader))

test(net)
```

As presented in the above cell, using both batch normalization and drop out for all the blocks and also fully connected layers has led to 0.99967 as accuracy. But a question rises that what would happen in not having all/some of batch normalizarion or drop out layers cases. The mentioned question is addressed below:

1st senario: 
- no batch normalization at all
- drop out for both blocks and fully connected layers

```python
net = ResNet(block_has_batchnorm=False, fc_has_batchnorm=False)
device = torch.device("cuda:0")
net.to(device)
record = train(net, 3)
```

```python
plt.plot(record["train"], label = "train loss")
plt.plot(record["validation"], label = "validation loss")
plt.legend()
plt.show()
```

```python
test(net)
```

As shown above, a signifcant decrease in network convergence has happened. It could have more effect also on the final result if the network was large enough. Accuracy is 0.999606.


2nd senario:

- batch normalization only for blocks
- drop out for both blocks and fully connected layers

```python
net = ResNet(block_has_batchnorm=True, fc_has_batchnorm=False)
device = torch.device("cuda:0")
net.to(device)
record = train(net, 3)
```

```python
plt.plot(record["train"], label = "train loss")
plt.plot(record["validation"], label = "validation loss")
plt.legend()
plt.show()
```

```python
test(net)
```

As shown above, a signifcant decrease in performance has happened, which demonstrates the rules of batch normalization in model power. Accuracy is 0.99731.

3rd senario:

- batch normalization for both blocks and fully connected layers
- no drop out for fully connected layers

```python
net = ResNet(fc_dropout_prob=0)
device = torch.device("cuda:0")
net.to(device)
record = train(net, 3)
```

```python
plt.plot(record["train"], label = "train loss")
plt.plot(record["validation"], label = "validation loss")
plt.legend()
plt.show()
```

```python
test(net)
```

As illustrated, generalization became poor and chance of overfitting increased. But since it only affected a single layer, it did not make much differences.

4th senario:

- batch normalization only for both blocks and fully connected layers
- no drop out for blocks

```python
net = ResNet(block_dropout_probs=[0, 0], fc_dropout_prob=0)
device = torch.device("cuda:0")
net.to(device)
record = train(net, 3)
```

```python
plt.plot(record["train"], label = "train loss")
plt.plot(record["validation"], label = "validation loss")
plt.legend()
plt.show()
```

```python
test(net)
```

As shown, it's the best model possible. Note that, this is the one suggested by the main ResNet paper. Because Convolution layers don't have much parameters and also cosidering their role in finding patterns and edges, drop out is not suitable for these layer types. Covergence rate and generalization are significantly improved.
