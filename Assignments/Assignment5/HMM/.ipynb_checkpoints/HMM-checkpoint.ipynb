```python
def calculate_beta(A, B):
    beta = np.array([[float(0) for _ in range(2)] for _ in range(time)]) #t, i
    beta[time - 1, :] += 1
    for t in range(len(beta) - 2, -1, -1):
        temp = 0
        for j in range(N):
            temp += beta[t + 1, j] * A[j, :] * B[j, Y[t + 1]]
        beta[t, :] += temp
    beta = (beta.T / beta.sum(axis=1)).T
    return beta
calculate_beta(A, B)
```

```python
alpha = calculate_alpha(A, B, pi)
beta = calculate_beta(A, B)
```

```python
def calculate_gamma(alpha, beta):
    gamma = np.array([[float(0) for _ in range(2)] for _ in range(time)]) #t, i
    gamma = alpha * beta
    gamma = (gamma.T / gamma.sum(axis=1)).T
    return gamma
calculate_gamma(alpha, beta)
```

```python
def calculate_kesi(alpha, beta, A, B):
    kesi = np.array([[[float(0) for _ in range(2)] for _ in range(2)] for _ in range(time - 1)]) #t, i, j
    for t in range(len(kesi)):
        for j in range(N):
            kesi[t, :, j] = alpha[t, :] * beta[t + 1, j] * A[j, :] * B[j, Y[t + 1]]
        kesi[t, :, :] /= kesi[t, :, :].sum()
    return kesi
calculate_kesi(alpha, beta, A, B)
```

```python
def update_B(Y, gamma):
    B = np.array([[random.uniform(0, 1) for _ in range(3)] for _ in range(2)]) # x, y
    temp = 0
    for k in range(3):
        temp += (((Y == k).astype(int) * gamma.T).T).sum(axis=0)
        B[:, k] = (((Y == k).astype(int) * gamma.T).T).sum(axis=0) / gamma.sum(axis=0)
    return B
update_B(Y, gamma)
```
